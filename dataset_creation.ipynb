```python
import pandas as pd 
import numpy as np
import ta
```

```python
df = pd.read_csv("data/data.csv")
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
# Ensure the column is treated as datetime
df['date'] = pd.to_datetime(df['date'])

df.drop(columns=["timestamp"])

# Find the oldest date
oldest_date = df['date'].min()
# Latest date
latest_date = df['date'].max()
print("Oldest date in the dataset:", oldest_date)
print("Latest date in the dataset:", latest_date)
```

```python
import pandas as pd

# Define the new start and end timestamps
start_date = pd.to_datetime("2015-10-08 14:00:00")
end_date = pd.to_datetime("2023-02-20 14:00:00")

# Filter df2 within this adjusted range
df_filtered = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Oldest and latest dates after filtering
oldest_date_df = df_filtered['date'].min()
latest_date_df = df_filtered['date'].max()
print("Oldest date in df2_filtered:", oldest_date_df)
print("Latest date in df2_filtered:", latest_date_df)
```

```python
df_filtered = df_filtered.set_index('date')
df_filtered = df_filtered.sort_index()

df_filtered.head()
```

```python
df_filtered.drop(columns=["timestamp", "symbol"], inplace=True)
```

```python
df2 = pd.read_csv("XAU_1m_data.csv", sep=';')
df2.columns = df2.columns.str.strip().str.lower()

# Convert 'date' column to datetime
df2['date'] = pd.to_datetime(df2['date'])

# Oldest and latest dates
oldest_date_df2 = df2['date'].min()
latest_date_df2 = df2['date'].max()
print("Oldest date in df2:", oldest_date_df2)
print("Latest date in df2:", latest_date_df2)
```

```python
# Define the date range from the first dataset
start_date = pd.to_datetime("2015-10-08 14:00:00")
end_date = pd.to_datetime("2023-02-20 14:00:00")

# Filter df2 within this range
df2_filtered = df2[(df2['date'] >= start_date) & (df2['date'] <= end_date)]

# Show result
df2_filtered.head()
# Oldest and latest dates
oldest_date_df2 = df2_filtered['date'].min()
latest_date_df2 = df2_filtered['date'].max()
print("Oldest date in df2:", oldest_date_df2)
print("Latest date in df2:", latest_date_df2)
```

```python
df2_gold = df2_filtered.copy().rename(columns={
    'open': 'open_gold',
    'close': 'close_gold',
    'low': 'low_gold',
    'high': 'high_gold',
    'volume': 'volume_gold'
})

df2_gold.head()
```

```python
# Set the 'date' column as the index
df2_gold = df2_gold.set_index('date')
df2_gold = df2_gold.sort_index()

df2_gold.head()
```

```python
# Merge on index (inner join to keep only matching timestamps)
merged_df = df_filtered.merge(df2_gold, left_index=True, right_index=True, how='inner')

merged_df.head()
```

```python
merged_df.isna().sum()
```

```python
ts = merged_df.copy()

# Simple Moving Average (SMA)
window_size = 20

# Calculate SMA
ts['SMA_20'] = ts['close'].rolling(window=window_size).mean()

# Fill missing values using backfill 
ts['SMA_20'] = ts['SMA_20'].bfill()
```

```python
#Exponential Moving Average
span = 20

# Calculate EMA
ts['EMA_20'] = ts['close'].ewm(span=span, adjust=False).mean()
```

```python
#Triple Exponential Moving Average
span = 20

# EMA of close
EMA = ts['close'].ewm(span=span, adjust=False).mean()

# EMA of EMA
EMA_of_EMA = EMA.ewm(span=span, adjust=False).mean()

# EMA of EMA of EMA
EMA_of_EMA_of_EMA = EMA_of_EMA.ewm(span=span, adjust=False).mean()

# Compute TEMA
ts['TEMA_20'] = 3*EMA - 3*EMA_of_EMA + EMA_of_EMA_of_EMA
```

```python
#Moving Average Convergence Divergence 
ema_fast = ts['EMA_20']
 # slow EMA
ema_slow = ts['close'].ewm(span=26, adjust=False).mean() 
ts['MACD'] = ema_fast - ema_slow
```

```python
# captures minute-to-minute price change
ts['returns'] = np.log(ts['close'] / ts['close'].shift(1))
```

```python
# RSI (14-period)
ts['RSI_14'] = ta.momentum.RSIIndicator(close=ts['close'], window=14).rsi()

# Bollinger Bands (20-period)
bb = ta.volatility.BollingerBands(close=ts['close'], window=20, window_dev=2)
ts['BB_upper'] = bb.bollinger_hband()
ts['BB_lower'] = bb.bollinger_lband()
ts['BB_width'] = ts['BB_upper'] - ts['BB_lower']

# ATR (14-period)
ts['ATR_14'] = ta.volatility.AverageTrueRange(high=ts['high'], low=ts['low'], close=ts['close'], window=14).average_true_range()

# VWAP (intraday, using cumulative volume-price)
ts['VWAP'] = (ts['close'] * ts['volume btc']).cumsum() / ts['volume btc'].cumsum()


# Rolling returns volatility (10-period)
ts['volatility_10'] = ts['close'].pct_change().rolling(10).std()

# High-Low range
ts['range'] = ts['high'] - ts['low']

# Lagged closes (1, 5, 10 minutes)
ts['close_lag1'] = ts['close'].shift(1)
ts['close_lag5'] = ts['close'].shift(5)
ts['close_lag10'] = ts['close'].shift(10)

# Lagged returns
ts['returns_lag1'] = ts['returns'].shift(1)
ts['returns_lag5'] = ts['returns'].shift(5)
ts['returns_lag10'] = ts['returns'].shift(10)

ts = ts.bfill()
```
